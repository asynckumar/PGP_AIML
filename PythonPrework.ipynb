```python
# Arithmetic operators
a = 10
b = 3

print("a =", a, "b =", b)
print("a + b  =", a + b,      " # +  addition")
print("a - b  =", a - b,      " # -  subtraction")
print("a * b  =", a * b,      " # *  multiplication")
print("a / b  =", a / b,      " # /  true division (float)")
print("a // b =", a // b,     " # // floor division (quotient, rounds down)")
print("a % b  =", a % b,      " # %  modulus (remainder)")
print("a ** b =", a ** b,     " # ** exponentiation (a to the power b)")
print()

# Arithmetic assignment operators
x = 10
print("Start x =", x)

x += 5
print("x += 5  ->", x, " # add and assign")

x -= 2
print("x -= 2  ->", x, " # subtract and assign")

x *= 3
print("x *= 3  ->", x, " # multiply and assign")

x /= 4
print("x /= 4  ->", x, " # divide and assign (float)")

x //= 2
print("x //= 2 ->", x, " # floor-divide and assign")

x %= 3
print("x %= 3  ->", x, " # modulus and assign")

x **= 2
print("x **= 2 ->", x, " # exponentiate and assign")
print()

# Comparison operators
price = 900
print("price =", price)

print("price == 900  ->", price == 900,   " # == equal to")
print("price != 800  ->", price != 800,   " # != not equal to")
print("price > 800   ->", price > 800,    " # >  greater than")
print("price < 1000  ->", price < 1000,   " # <  less than")
print("price >= 900  ->", price >= 900,   " # >= greater than or equal to")
print("price <= 1000 ->", price <= 1000,  " # <= less than or equal to")
print()

# Logical operators
available = True
in_budget = (price <= 1000)   # comparison -> boolean

print("available =", available)
print("in_budget =", in_budget, "(price <= 1000)")

print("in_budget and available ->", in_budget and available,
      " # and  True only if both are True")

print("in_budget or available  ->", in_budget or available,
      " # or   True if at least one is True")

print("not available          ->", not available,
      " # not  flips True/False")
```
